Preprocessing main step

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
import pandas as pd

def preprocess_data(df, df_test):
    # Calculate the threshold for missing values
    missing_threshold = 0.4

    # Determine columns with more than 40% missing values
    missing_perc = df_test.isnull().mean()
    cols_to_drop = missing_perc[missing_perc >= missing_threshold].index.tolist()
    print(f"Columns having {missing_perc} with more than {missing_threshold*100}% missing values: {cols_to_drop}")

    # Drop these columns from both datasets
    df = df.drop(columns=cols_to_drop, errors='ignore')
    df_test = df_test.drop(columns=cols_to_drop, errors='ignore')

    # Identify numerical and categorical columns
    numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
    categorical_cols = df.select_dtypes(include=['object']).columns.tolist()

    # Remove the target variable 'SalePrice' from the list of numerical columns
    if 'SalePrice' in numerical_cols:
        numerical_cols.remove('SalePrice')

    # Define preprocessing for numerical and categorical data
    numerical_transformer = Pipeline(steps=[
        ('imputer', SimpleImputer(strategy='mean')),
        ('scaler', StandardScaler())
    ])

    categorical_transformer = Pipeline(steps=[
        ('imputer', SimpleImputer(strategy='most_frequent')),
        ('onehot', OneHotEncoder(handle_unknown='ignore'))
    ])

    # Combine numerical and categorical transformers
    preprocessor = ColumnTransformer(
        transformers=[
            ('num', numerical_transformer, numerical_cols),
            ('cat', categorical_transformer, categorical_cols)
        ]
    )

    # Separate features and target
    X = df.drop('SalePrice', axis=1)
    y = df['SalePrice']
    X_test = df_test.copy()

    # Fit the preprocessor on the training data and transform both training and test data
    X_preprocessed = preprocessor.fit_transform(X)
    X_test_preprocessed = preprocessor.transform(X_test)

    return X_preprocessed, y, X_test_preprocessed
```

Another way of preprocessing

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
import pandas as pd

def preprocess_data1(df, df_test, max_unique_values=10):
    # Calculate the threshold for missing values
    missing_threshold = 0.4

    # Determine columns with more than 40% missing values
    missing_perc = df_test.isnull().mean()
    cols_to_drop = missing_perc[missing_perc >= missing_threshold].index.tolist()
    print(f"Columns with more than {missing_threshold*100}% missing values: {cols_to_drop}")

    # Drop these columns from both datasets
    df = df.drop(columns=cols_to_drop, errors='ignore')
    df_test = df_test.drop(columns=cols_to_drop, errors='ignore')

    # Identify numerical and categorical columns
    numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
    categorical_cols = df.select_dtypes(include=['object']).columns.tolist()

    # Analyze numerical columns to identify potential categorical columns
    for col in numerical_cols:
        if df[col].nunique() <= max_unique_values:
            print(f"Numerical column '{col}' is treated as categorical because it has {df[col].nunique()} unique values.")
            categorical_cols.append(col)

    # Remove the target variable 'SalePrice' from the list of numerical columns
    if 'SalePrice' in numerical_cols:
        numerical_cols.remove('SalePrice')

    # Remove columns that are being treated as categorical from numerical columns
    numerical_cols = [col for col in numerical_cols if col not in categorical_cols]

    # Define preprocessing for numerical and categorical data
    numerical_transformer = Pipeline(steps=[
        ('imputer', SimpleImputer(strategy='mean')),
        ('scaler', StandardScaler())
    ])

    categorical_transformer = Pipeline(steps=[
        ('imputer', SimpleImputer(strategy='most_frequent')),
        ('onehot', OneHotEncoder(handle_unknown='ignore'))
    ])

    # Combine numerical and categorical transformers
    preprocessor = ColumnTransformer(
        transformers=[
            ('num', numerical_transformer, numerical_cols),
            ('cat', categorical_transformer, categorical_cols)
        ]
    )

    # Separate features and target
    X = df.drop('SalePrice', axis=1)
    y = df['SalePrice']
    X_test = df_test.copy()

    # Fit the preprocessor on the training data and transform both training and test data
    X_preprocessed = preprocessor.fit_transform(X)
    X_test_preprocessed = preprocessor.transform(X_test)

    return X_preprocessed, y, X_test_preprocessed
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import ElasticNet
from sklearn.metrics import mean_squared_error
import pandas as pd

# Function to train and predict using Random Forest Regressor
def random_forest_regressor(X, y, X_test, df_test, id_column='Id', y_test=None, n_estimators=100, max_depth=None, random_state=42):
    rf_model = RandomForestRegressor(n_estimators=n_estimators, max_depth=max_depth, random_state=random_state)
    rf_model.fit(X, y)
    rf_predictions = rf_model.predict(X_test)


    rf_results = pd.DataFrame({id_column: df_test[id_column], 'SalePrice': rf_predictions})
    return rf_model, rf_results

# Function to train and predict using Elastic Net
def elastic_net_regressor(X, y, X_test, df_test, id_column='Id', y_test=None, alpha=1.0, l1_ratio=0.5, random_state=42):
    enet_model = ElasticNet(alpha=alpha, l1_ratio=l1_ratio, random_state=random_state)
    enet_model.fit(X, y)
    enet_predictions = enet_model.predict(X_test)

    enet_results = pd.DataFrame({id_column: df_test[id_column], 'SalePrice': enet_predictions})
    return enet_model, enet_results

```

```python
from xgboost import XGBRegressor
import pandas as pd
from sklearn.metrics import mean_squared_error

def xgboost_regressor(X, y, X_test, df_test, id_column='Id', n_estimators=100, max_depth=None, random_state=42):
    # Initialize and train XGBoost model
    xgb_model = XGBRegressor(n_estimators=n_estimators, max_depth=max_depth, random_state=random_state, objective='reg:squarederror')
    xgb_model.fit(X, y)

    # Make predictions
    xgb_predictions = xgb_model.predict(X_test)

    # Create a DataFrame with test IDs and predictions
    xgb_results = pd.DataFrame({
        id_column: df_test[id_column],
        'SalePrice': xgb_predictions
    })

    # Save predictions to CSV file
    xgb_results.to_csv('submission.csv', index=False)
    print("Predictions saved.")

    return xgb_model, xgb_results
```

```python
import tensorflow_decision_forests as tfdf
import pandas as pd
import tensorflow as tf
import scipy.sparse as sp

def tfdf_regressor(X, y, X_test, df_test, id_column='Id', num_trees=100, batch_size=32):
    # Convert sparse matrices to dense arrays if necessary
    if isinstance(X, (sp.csr_matrix, sp.csc_matrix)):
        X = X.toarray()
    if isinstance(X_test, (sp.csr_matrix, sp.csc_matrix)):
        X_test = X_test.toarray()

    # Convert DataFrames to ensure feature names are valid
    X_df = pd.DataFrame(X)
    X_test_df = pd.DataFrame(X_test)

    # Rename columns to ensure no empty names
    X_df.columns = [f'feature_{i}' for i in range(X_df.shape[1])]
    X_test_df.columns = [f'feature_{i}' for i in range(X_test_df.shape[1])]

    # Convert DataFrames to dictionaries of features
    train_features = {col: tf.convert_to_tensor(X_df[col].values, dtype=tf.float32) for col in X_df.columns}
    test_features = {col: tf.convert_to_tensor(X_test_df[col].values, dtype=tf.float32) for col in X_test_df.columns}

    # Convert y to TensorFlow tensor
    y_tensor = tf.convert_to_tensor(y.values, dtype=tf.float32)

    # Convert DataFrames to TensorFlow Datasets
    train_data = tf.data.Dataset.from_tensor_slices((train_features, y_tensor))
    test_data = tf.data.Dataset.from_tensor_slices(test_features)

    # Batch the datasets
    train_data = train_data.batch(batch_size)
    test_data = test_data.batch(batch_size)

    # Create a TF-DF RandomForest model
    model = tfdf.keras.RandomForestModel(task=tfdf.keras.Task.REGRESSION, num_trees=num_trees)

    # Train the model
    model.fit(train_data)

    # Make predictions
    predictions = model.predict(test_data)

    # Convert predictions to a DataFrame
    predictions_df = pd.DataFrame({
        id_column: df_test[id_column],
        'SalePrice': predictions.flatten()  # Flatten the predictions if necessary
    })

    # Save predictions to CSV file
    predictions_df.to_csv('submission.csv', index=False)
    print("Predictions saved.")

    return model, predictions_df
```

```python
model, results = tfdf_regressor(X, y, X_test, df_test)
```

```python
import matplotlib.pyplot as plt

def plot_feature_importances(model):
    """Plots feature importances for a TensorFlow Decision Forest model."""

    # Get the inspector
    inspector = model.make_inspector()

    # Try to use inspector.importance()
    try:
        importances = inspector.importance()
        feature_names = importances['feature_names']
        importances = importances['importances']

        # Create a DataFrame for plotting
        feature_importances_df = pd.DataFrame({
            'Feature': feature_names,
            'Importance': importances
        })

        # Sort features by importance
        feature_importances_df = feature_importances_df.sort_values(by='Importance', ascending=False)


        # Plot
        plt.figure(figsize=(12, 8))
        plt.barh(feature_importances_df['Feature'], feature_importances_df['Importance'])
        plt.xlabel('Importance')
        plt.title('Feature Importances')
        plt.gca().invert_yaxis()  # Top features on top
        plt.show()

    except AttributeError:
        print("inspector.importance() not available. Consider alternative methods.")

        # Example using tfdf.model_plotter
        tfdf.model_plotter.plot_model(model)
        plt.show()
```
